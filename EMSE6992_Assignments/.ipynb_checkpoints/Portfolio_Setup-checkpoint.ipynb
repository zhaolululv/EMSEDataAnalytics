## Portfolios and Setup

Below are instructions on how to install, edit, and create a Portfolio. The portfolio template that you will be downloading is a template that I created and which was donated to us by Karl Broman.  All of the pages are in markdown which is really similar to html so editing the files is trivial.

In order to receive full credit for assignments and final project, you must install the template, personalize the protfolio (entering in your bio/CV/publications/etc.) and send me the link to the newly created portfolio (username.github.io). Not looking for anything perfect here, just want to see your portfolio successfully hosted on Github and want you to show progress towards getting the portfolio to a point that will enable you to build upon in the future.

Download portfolio template from Github (save to your desktop)
https://github.com/bsharve/bsharve.github.io

Create a github account
https://github.com/

Within GitHub...Create a repository with your username in the within the name of the repository for example: "username.github.io" 

Upload bsharve.github.io zip file to the new username.github.io repository

Download and Install Anaconda
https://www.anaconda.com/download/

Download and Install Github Desktop 
https://desktop.github.com/

Download and Install Atom IDE
https://atom.io/

Go to https://username.github.io/ to check and see if the github installation worked correctly

There are two ways to edit the portfolios:
 - Directly within github
 - Using Github Desktop
 
To edit the portfolio directly within github, go to your https://github.com/username/username.github.io and click directly on the files that you would like to edit.

To edit the portolio using Github Desktop, once you have downloaded and installed Github Desktop, go to the file menu and select "clone repository."   This will enable you to clone the Github repository "username/username.github.io" to your local computer (make sure to keep track of what directory you cloned the repository to). 

After you clone the repository you can either use the Atome IDE or any other IDE to edit the files. 

Within Atom, after you edit a file in the cloned repository, the Github IDE automatically tracks changes that are made. Once changes to the local repository are made, you must push those changes to the repository hosted on Github.  

In order to do this, after you edit a file, go back to the Github Desktop and on the left had side of the application you should see "changed files."  This is where the Github Desktop application has been tracking changes made to the local repository.  

The next step after changes have been made is to push them to the repository hosted on Github. To do this, you must add a description of the edits to the "Summary" text box located on the left lower side of the Github Desktop application and the press the "Commit to Master" button on the lower left hand side of the Github Desktop application.

After you push the "Commit to Master" button, you will now have to push the "Push Origin" button on the top right hand side of Github Desktop.

This will successfully enable you to update the files from your local repository to the hosted repository on Github.

You can also edit and conduct version control directly from the command line.  You will have to install Git from the command line and use the following useful github command line comands:

git init git branch -m master gh-pages git add --all git status git commit -m "first commit" git remote add origin https://github.com/bsharve/bsharve.github.io.git //git remote remove origin git push -u origin gh-pages
